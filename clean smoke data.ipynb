```python
import numpy as np
import pandas as pd
import os
import math
from sympy import *
```

```python
filename = 'smoking_driking_dataset_Ver01.csv'
data = pd.read_csv(filename)

raw_data = data.values

attriubutes = np.asarray(data.columns)

print("These are the features", attriubutes)
```

```python
data[:10]
```

```python
cols = range(0,24) # 24 columns in the data set
data = raw_data[:,cols] # defining the data as a matrix
```

```python
# Replacing strings with numbers. Starting with gender:
data[data == 'Female'] = 0
data[data == 'Male'] = 1

# Replacing Y = yes with 1 and N = no with 0 in regards to drinking
data[data == 'Y'] = 1
data[data == 'N'] = 0
```

```python
# Now I standardize the data, but not the columns that are ordinal of character 
# and not the binary variables.
# The one ordinal feature is the degree of smoking habits.
# The binary features are the sex and the drinking habits.



# Standardize the data, execept last two columns
for i in range(0,22):
    data[:,i] = (data[:,i] - data[:,i].mean()) / data[:,i].std()
# Here everything is standardized, except the last two columns, 

# Now I will only acces the last column and standardize it:
data[:,23] = (data[:,22] - data[:,22].mean()) / data[:,22].std()
```

```python
# Now I will use one-out-of-K encoding for the ordinal feature, smoking degree:
smoking = data[:,22]

encoding = np.zeros((len(smoking), 3)) # three different smoking stages

for i in range(0,len(smoking)):
    for j in range(0,3):
        if smoking[i] == j+1:
            encoding[i,j] = 1
        else:
            encoding[i,j] = 0

# Combining the data and the encoded data:
data = np.concatenate((data[:,0:22], encoding), axis = 1)
```
